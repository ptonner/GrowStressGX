```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
meta = pd.read_excel("metadata08222013.xls")
meta.head()
```

```python
# has_growth = meta['condition.descriptor'].str.contains("density")
has_growth = meta['condition'].str.contains("OD600")
has_growth = has_growth.replace(np.nan,False)
```

```python
meta[has_growth].condition.unique()
```

```python
"%d total conditions, %d unique" % (sum(has_growth),meta[has_growth].condition.unique().shape[0])
```

```python
meta.columns
```

```python
meta_growth = meta[has_growth]

meta_facciotti = meta[meta['literature.ref']=='Facciotti et al., 2010, BMC Systems Biology 4:64']
meta_facciotti = meta_facciotti[meta_facciotti['Column.name'].str.find('X20060818') != 0]
meta_facciotti = pd.DataFrame([(x[0],x[1],int(x[2]),float(x[3]),int(x[4][-1])) for x in meta_facciotti['Column.name'].str.split("_").tolist()],
                              columns = ['strain','media','time','od','replicate'],
                              index=meta_facciotti.index)
meta_facciotti.time = 1.*meta_facciotti.time/60
meta_facciotti['log_od'] = np.log2(meta_facciotti.od)
meta_facciotti.to_csv("facciotti_meta.csv",index=True)
```

```python
plt.scatter(meta_facciotti.time,meta_facciotti.log_od)
```

```python
g = meta_facciotti.groupby(['strain','media','replicate'])
```

```python
g.plot(x='time',y='log_od',kind='scatter')
```
