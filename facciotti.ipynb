```python
def gpDerivative(x,gp,k=0):

	if x.ndim == 1:
		x = x[:,np.newaxis]

	# from Solak et al.
	mu,ignore = gp.predictive_gradients(x)
	mu = mu[:,k]
	ignore,cov = gp.predict(x,full_cov=True)
	mult = [[((1./gp.kern.lengthscale[k])*(1-(1./gp.kern.lengthscale[k])*(y - z)**2)) for y in x[:,k]] for z in x[:,k]]
	return mu, mult*cov
```

```python
selected = []
residuals = []

for i in range(10):
    pvals = []
    genes = []
    for c in x.columns:
        if c in selected:
            continue
        a = x[selected+[c]].values
        a = sm.add_constant(a)
        est = sm.OLS(y, a)
        est = est.fit()
        pvals.append(est.pvalues[-1])
        genes.append(c)

    pvals = pd.DataFrame(pvals,columns=['pvalue'])
    pvals['gene'] = genes
    pvals = pvals.sort_values('pvalue',ascending=True)
    selected.append(pvals.iloc[0,:]['gene'])
    
    a = x[selected].values
    a = sm.add_constant(a)
    est = sm.OLS(y, a)
    est = est.fit()
    residuals.append(est.resid.sum())
```

```python
f_output1 = lambda x: 4. * np.cos(x/5.) - .4*x - 35. + np.random.rand(x.size)[:,None] * 2.
f_output2 = lambda x: 6. * np.cos(x/5.) + .2*x + 35. + np.random.rand(x.size)[:,None] * 8.


#{X,Y} training set for each output
X1 = np.random.rand(100)[:,None]; X1=X1*75
X2 = np.random.rand(100)[:,None]; X2=X2*70 + 30
Y1 = f_output1(X1)
Y2 = f_output2(X2)
#{X,Y} test set for each output
Xt1 = np.random.rand(100)[:,None]*100
Xt2 = np.random.rand(100)[:,None]*100
Yt1 = f_output1(Xt1)
Yt2 = f_output2(Xt2)
```

```python
xlim = (0,100); ylim = (0,50)
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
ax1.set_xlim(xlim)
ax1.set_title('Output 1')
ax1.plot(X1[:,:1],Y1,'kx',mew=1.5,label='Train set')
ax1.plot(Xt1[:,:1],Yt1,'rx',mew=1.5,label='Test set')
ax1.legend()
ax2 = fig.add_subplot(212)
ax2.set_xlim(xlim)
ax2.set_title('Output 2')
ax2.plot(X2[:,:1],Y2,'kx',mew=1.5,label='Train set')
ax2.plot(Xt2[:,:1],Yt2,'rx',mew=1.5,label='Test set')
ax2.legend()
```

```python
def plot_2outputs(m,xlim,ylim):
    fig = plt.figure(figsize=(12,8))
    #Output 1
    ax1 = fig.add_subplot(211)
    ax1.set_xlim(xlim)
    ax1.set_title('Output 1')
    m.plot(plot_limits=xlim,fixed_inputs=[(1,0)],which_data_rows=slice(0,100),ax=ax1)
    ax1.plot(Xt1[:,:1],Yt1,'rx',mew=1.5)
    #Output 2
    ax2 = fig.add_subplot(212)
    ax2.set_xlim(xlim)
    ax2.set_title('Output 2')
    m.plot(plot_limits=xlim,fixed_inputs=[(1,1)],which_data_rows=slice(100,200),ax=ax2)
    ax2.plot(Xt2[:,:1],Yt2,'rx',mew=1.5)
```
